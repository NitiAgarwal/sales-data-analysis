```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. LOAD DATA (replace with your CSV if needed)
# -----------------------------
data = {
    'Date': ['01-01-2024','02-01-2024','03-01-2024','04-01-2024','05-01-2024',
             '06-02-2024','07-02-2024','08-03-2024','09-03-2024','10-04-2024'],
    'Product': ['Laptop','Mobile','Tablet','Laptop','Mobile',
                'Laptop','Tablet','Mobile','Laptop','Tablet'],
    'Region': ['North','South','East','West','North',
               'South','East','West','North','South'],
    'Sales_Amount': [50000,30000,15000,55000,28000,60000,20000,35000,45000,25000],
    'Quantity': [2,5,3,2,4,3,2,5,2,3]
}

df = pd.DataFrame(data)

# -----------------------------
# 2. BASIC INFO
# -----------------------------
print("\nFirst 5 Rows:\n", df.head())
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:\n", df.isnull().sum())

print("\nDescriptive Statistics:\n", df.describe())

# -----------------------------
# 3. DATE CONVERSION + MONTH
# -----------------------------
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month_name()

# -----------------------------
# 4. PRODUCT WISE SALES
# -----------------------------
product_sales = df.groupby('Product')['Sales_Amount'].sum()
print("\nTop Selling Products:\n", product_sales)

product_sales.plot(kind='bar')
plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Sales Amount")
plt.show()

# -----------------------------
# 5. REGION WISE SALES
# -----------------------------
region_sales = df.groupby('Region')['Sales_Amount'].sum()
print("\nRegion Wise Sales:\n", region_sales)

region_sales.plot(kind='bar')
plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales Amount")
plt.show()

# -----------------------------
# 6. MONTHLY SALES
# -----------------------------
monthly_sales = df.groupby('Month')['Sales_Amount'].sum()
print("\nMonthly Sales:\n", monthly_sales)

monthly_sales.plot(kind='bar')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.show()

# -----------------------------
# 7. INSIGHTS
# -----------------------------
print("\nINSIGHTS:")
print("1. Laptop generates highest revenue")
print("2. North region contributes strong sales")
print("3. January shows high sales activity")

# -----------------------------
# 8. SAVE CLEAN DATA
# -----------------------------
df.to_csv("cleaned_sales_data.csv", index=False)

print("\nProject Completed Successfully 🚀")
```
